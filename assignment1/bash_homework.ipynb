Question 4a: Using just bash commands can you find out what is the average length of protein in this strain?

Step 1. Count the number of proteins in the proteome.

Grep the seq info lines "^>" and count the number of lines (wc -l). 

```bash
cat NC_000913.faa | grep "^>" | wc -l
```

Step 2. Count the number of amino acids in the proteome. 

Reverse the grep to drop seq info lines (-v "^>") and translate out carriage returns (tr -d "\n").  

```bash
cat NC_000913.faa | grep -v "^>" | tr -d "\n"
```

Cound the amino acid characters (wc -m).

```bash
cat NC_000913.faa | grep -v "^>" | tr -d "\n" | wc -m
```

```bash
echo $(cat NC_000913.faa | grep -v "^>" | tr -d "\n" | wc -m)/$(cat NC_000913.faa | grep "^>" | wc -l) | bc
```
